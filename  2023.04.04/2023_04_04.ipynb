```python
invalid_number = ns_book4['발행년도'].str.contains('\D', na=True)
print(invalid_number.sum())
ns_book4[invalid_number].head()
```

```python
ns_book5 = ns_book4.replace({'발행년도':'.*(\d{4}).*'}, r'\1', regex=True)
ns_book5[invalid_number].head()
```

```python
unkown_year = ns_book5['발행년도'].str.contains('\D', na=True) # 숫자 이외의 문자가 들어간 행의 개수와 데이터
print(unkown_year.sum())
ns_book4[unkown_year].head()  
```

```python
ns_book5.loc[unkown_year, '발행년도'] = '-1'
ns_book5 = ns_book5.astype({'발행년도': 'int32'})
ns_book5['발행년도'].gt(4000).sum() # 4000 보다 큰 값의 개수
```

```python
(ns_book5['발행년도'] > 4000).sum()
```

```python
dangun_yy_rows = ns_book5['발행년도'].gt(4000) # 4000본다 큰 값 찾기
ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows, '발행년도'] - 2333 # 찾은 값에서 2333을 빼서 서기로 바꾸기
dangun_year = ns_book5['발행년도'].gt(4000) # 다시 4000 보다 큰 값을 찾기
print(dangun_year.sum())
ns_book5[dangun_year].head(2)
```

```python
ns_book5.loc[dangun_year, '발행년도'] = -1 # 연도가 이상하게 높은 도서의 발행년도를 -1로 지정
```

```python
old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
print(old_books.sum())
ns_book5[old_books]
```

```python
import requests
from bs4 import BeautifulSoup
def get_book_title(isbn):
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  return title
```

```python
def get_book_author(isbn):
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  author = soup.find('span', attrs={'class':'info_auth'}).find_all('a')
  result = list(tag.get_text() for tag in author)
  return {'제목':title,'저자':result}
```
