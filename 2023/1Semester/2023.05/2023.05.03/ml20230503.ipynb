# 도미와 빙어를 구분하는 인공지능

### 데이터 준비하기

##### 도미 데이터 준비하기

```python
# 생선의 길이
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 
                31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 
                34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 
                38.5, 38.5, 39.5, 41.0, 41.0]

# 생선의 무게
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 
                475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 
                575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 
                920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
import matplotlib.pyplot as plt

# 산점도 그리기
plt.scatter(bream_length, bream_weight)
plt.title('Sea bream')
plt.xlabel('length')
plt.ylabel('weigth')
plt.show()
```

##### 빙어 데이터 준비하기

```python
# 생선의 길이
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 
                12.4, 13.0, 14.3, 15.0]

# 생선의 무게
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 
                13.4, 12.2, 19.7, 19.9]
```

```python
import matplotlib.pyplot as plt

# 산점도 그리기
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.title('Sea bream & smelt')
plt.xlabel('length')
plt.ylabel('weigth')
plt.show()
```

### 데이터 합치기

##### 도미 데이터와 빙어 데이터 합치기

```python
length = bream_length + smelt_length # 도미의 길이 데이터와 빙어의 길이 데이터 합치기
weigth = bream_weight + smelt_weight # 도미의 무게 데이터와 빙어의 무게 데이터 합치기
print(f'length : {length}\nweigth : {weigth}')
```

##### 2차원로 변환하기

```python
# 1차원 리스트인 생선의 길이 데이터와 생선의 무게 데이터를 zip() 함수와 리스트 내포 구문을 사용하여 2차원 리스트 만들기
# zip()는 리스트에서 원소를 하나씩 거내 반환
fish_data = [[l, w] for l, w in zip(length, weigth)] # 리스트 내포 구문을 사용하여 fish_data 변수에 결과값인 2차원 리스트를 대입
fish_data
```

### 테스트 해보기

##### 1과 0으로 된 리스트 만들기

```python
fish_target = [1] * 35 + [0] * 14 
# 1 : 도미, 0 : 빙어
print(fish_target)
```

##### 인공지능 학습하기

```python
from sklearn.neighbors import KNeighborsClassifier
# 사이킷런 패키지에 k-최근접 이웃 알고리즘 구현 클래스인 KNeighborsClassifier 임포트

kn = KNeighborsClassifier() # KNeighborsClassifier 객체 생성
kn.fit(fish_data, fish_target) # 데이터 학습
kn.score(fish_data, fish_target) # 산출
```

##### 새로운 데이터 입력하기

```python
kn.predict([[30, 600]])
# predict() : 새로운 데이터의 정답을 예측
```

##### kn49모델로 학습하기

```python
kn49 = KNeighborsClassifier(n_neighbors=49) # 참고 데이터를 49개로 한 kn49 모델
kn49.fit(fish_data, fish_target) # 데이터 학습
kn49.score(fish_data, fish_target) # 산출
```

### 훈련 데이터와 테스트 데이터 생성하기

##### 데이터 준비하기

```python
# 생선의 길이
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

# 생선의 무게
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

##### 2차원 리스트 만들기

```python
fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)]
fish_target = [1] * 35 + [0] * 14
```

##### 훈련데이터와 테스트데이터를 나눠서 인공지능 학습하기

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

# 훈련 세트로 입렷값(index : 0 ~ 34)
train_input = fish_data[:35]

# 훈련 세트로 결과값(index : 0 ~ 34)
train_target = fish_target[:35]

# 테스트 세트로 입렷값(index : 35 ~)
test_input = fish_data[35:]

# 테스트 세트로 결과값(index : 35 ~)
test_target = fish_target[35:]

kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

### 넘파이 사용하기

```python
import numpy as np # 넘파이 클래스 임포트

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

print('input_arr.shape :', input_arr.shape)
```

##### 데이터 섞기

```python
np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index) # 배열을 무작위로 섞기

# 훈련 데이터
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

# 테스트 데이터
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

##### 산점도 그리기

```python
import matplotlib.pyplot as plt

# 산점도 그리기
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
plt.title('train data & test data')
plt.xlabel('length')
plt.ylabel('weigth')
plt.show()
```
