```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
df = pd.read_csv('학생별과목별성적.csv', encoding='CP949')
df1 = df[df['이름'] == '박지영']
df2 = df[df['이름'] == '김태함']
df3 = df[df['이름'] == '권혁진']
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,10))
fig.set_facecolor('white')
ax = fig.add_subplot()
ax.plot(df1['과목'], df1['점수'], label='박지영', color='red')
ax.plot(df2['과목'], df2['점수'], label='김태함', color='blue')
ax.plot(df3['과목'], df3['점수'], label='권혁진', color='purple')
plt.title('학생별 과목별 성적')
plt.xlabel('과목')
plt.ylabel('점수(단위 : 점)')
plt.legend(loc='lower left')
plt.show()
```
