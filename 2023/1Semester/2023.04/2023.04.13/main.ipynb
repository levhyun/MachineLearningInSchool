```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig=plt.figure()
ax=fig.add_subplot(1,1,1)   
labels=['eng','math']
x=[3,4]
y=[1,2]
ax.bar(x,y, tick_label=labels)
plt.show()
```
