```python
import requests
from bs4 import BeautifulSoup
def get_book_title(isbn):
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  return title
```

```python
def get_book_author(isbn):
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  author = soup.find('span', attrs={'class':'info_auth'}).find_all('a')
  result = list(tag.get_text() for tag in author)
  return {'제목':title,'저자':result}
```

```python
def get_book_pub(isbn): # 책의 출판사를 반환하는 함수
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}' # URL 포맷.

  r = requests.get(url) # URL에 GET 요청.

  html = BeautifulSoup(r.text, 'html.parser') # URL에 GET 요청을 보낸 결과의 HTML을 파싱.

  title = html.find('a', attrs={'class':'gd_name'}).get_text() # html에 a태그 중 클래스이름이 gd_name인 태그의 텍스트값을 반환.

  author = html.find('span', attrs={'class':'info_auth'}).find_all('a') # html에 span태그 중 클래스이름이 info_auth인 태그 안에 모든 a태그를 반환.
  result = list(tag.get_text() for tag in author) # a태그의 텍스트값을 추출하여 리스트화.

  pub = html.find('span', attrs={'class':'info_pub'}).find('a').get_text() # html에 span태그 중 클래스이름이 info_pub인 태그 안에 a태그의 텍스트값을 반환.

  return {'제목':title,'저자':result, '출판사':pub} # 책제목, 책저자, 책출판사를 딕셔너리로 반환.
```

```python
import re
def get_book_year(isbn): # 책의 출판사를 반환하는 함수
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}' # URL 포맷.

  r = requests.get(url) # URL에 GET 요청.

  html = BeautifulSoup(r.text, 'html.parser') # URL에 GET 요청을 보낸 결과의 HTML을 파싱.

  title = html.find('a', attrs={'class':'gd_name'}).get_text() # html에 a태그 중 클래스이름이 gd_name인 태그의 텍스트값을 반환.

  author = html.find('span', attrs={'class':'info_auth'}).find_all('a') # html에 span태그 중 클래스이름이 info_auth인 태그 안에 모든 a태그를 반환.
  result = list(tag.get_text() for tag in author) # a태그의 텍스트값을 추출하여 리스트화.

  pub = html.find('span', attrs={'class':'info_pub'}).find('a').get_text() # html에 span태그 중 클래스이름이 info_pub인 태그 안에 a태그의 텍스트값을 반환.

  # year = html.find('span', attrs={'class':'info_date'}).get_text()[:4] # html에 span태그 중 클래스이름이 info_date 태그의 텍스트값의 인덱스번호 0부터 3까지의 값만 반환.
  year_str = html.find('span', attrs={'class':'info_date'}).get_text() # html에 span태그 중 클래스이름이 info_date 태그의 텍스트값을 반환.
  year = re.findall(r'\d{4}', year_str) # 정규표현식으로 앞의 4자리만 추출

  return {'제목':title,'저자':result, '출판사':pub, '발행년도':year[0]} # 책제목, 책저자, 책출판사, 책발행년도를 딕셔너리로 반환.
```

```python
def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']
  
  isbn = row['ISBN']
  url = f'http://www.yes24.com/Product/Search?domain=BOOK&query={isbn}' 

  r = requests.get(url)

  html = BeautifulSoup(r.text, 'html.parser')

  try:
    if pd.isna(title):
      title = html.find('a', attrs={'class':'gd_name'}).get_text()
  except AttributeError:
    pass

  try:
    if pd.isna(author):
      authors = html.find('span', attrs={'class':'info_auth'}).find_all('a') 
      author = list(tag.get_text() for tag in authors) 
  except AttributeError:
    pass

  try:
    if pd.isna(pub):
      pub = html.find('span', attrs={'class':'info_pub'}).find('a').get_text()
  except AttributeError:
    pass
  
  try:
    if pd.isna(year):
      year_str = html.find('span', attrs={'class':'info_date'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass
  
  return title,author,pub,year
```
