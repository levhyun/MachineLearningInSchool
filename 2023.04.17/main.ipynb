```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

df = pd.read_csv('1군전염병발병현황_년도별.csv', encoding='EUC-KR')
df
```

```python
plt.subplots(figsize=(12,6))
plt.style.use('ggplot')
barWidth = 0.1
index = np.arange(df.drop_duplicates(['년도'])['년도'].count())
b1 = plt.bar(index, df['콜레라'], barWidth, alpha=0.4, color='red', label='콜레라')
b2 = plt.bar(index + barWidth, df['장티푸스'], barWidth, alpha=0.4, color='blue', label='장티푸스')
b3 = plt.bar(index + 2 * barWidth, df['이질'], barWidth, alpha=0.4, color='purple', label='이질')
b4 = plt.bar(index + 3 * barWidth, df['대장균'], barWidth, alpha=0.4, color='green', label='대장균')
plt.xticks(index, df['년도'])
plt.ylabel('발생건수', size = 9)
plt.legend()
plt.show()
```
